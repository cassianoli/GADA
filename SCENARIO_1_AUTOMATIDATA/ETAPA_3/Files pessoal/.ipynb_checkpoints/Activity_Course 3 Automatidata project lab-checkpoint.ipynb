## BONUS CONTENT

To confirm your conclusion, consider the following experiment:
1. Create a sample of coordinates from a normal distribution&mdash;in this case 1,500 pairs of points from a normal distribution with a mean of 10 and a standard deviation of 5
2. Calculate the distance between each pair of coordinates 
3. Group the coordinates by endpoint and calculate the mean distance between that endpoint and all other points it was paired with
4. Plot the mean distance for each unique endpoint

```python
#1. Generate random points on a 2D plane from a normal distribution

# Define the number of points to generate
n = 1500

# Generate x and y coordinates from a normal distribution (mean = 10, standard deviation = 5)
x = np.random.normal(10, 5, n)
y = np.random.normal(10, 5, n)

# Combine x and y into a 2D array
points = np.column_stack((x, y))

# Create a scatter plot to visualize the points
plt.figure(figsize=(6,6))
plt.scatter(x, y, alpha=0.6)
plt.title("Random Points from Normal Distribution (mean=10, std=5)")
plt.xlabel("X coordinate")
plt.ylabel("Y coordinate")
plt.show()
```

```python
# 2. Calculate Euclidean distances between points in first half and second half of array

# Split the array into two halves
first_half = points[:len(points)//2]
second_half = points[len(points)//2:]

# Calculate Euclidean distances between each pair of points (first vs second half)
distances = np.sqrt(np.sum((first_half - second_half)**2, axis=1))

# Display the first few distances to verify
print(distances[:10])
```

```python
# 3. Group the coordinates by "drop-off location", compute mean distance

# Create a DataFrame to organize the data
df_distances = pd.DataFrame({
    'drop_off_x': second_half[:, 0],
    'drop_off_y': second_half[:, 1],
    'distance': distances
})

# Group by "drop-off location" (x, y) and calculate the mean distance
mean_distance_by_dropoff = (
    df_distances
    .groupby(['drop_off_x', 'drop_off_y'])['distance']
    .mean()
    .reset_index()
)

# Display the first few results
print(mean_distance_by_dropoff.head())
```

```python
# 4. Plot the mean distance between each endpoint ("drop-off location") and all points it connected to

# Sort values by mean distance (ascending order)
mean_distance_by_dropoff_sorted = mean_distance_by_dropoff.sort_values(by='distance', ascending=True)

# Create a bar plot
plt.figure(figsize=(10,6))
plt.bar(range(len(mean_distance_by_dropoff_sorted)), mean_distance_by_dropoff_sorted['distance'])
plt.title('Mean Distance by Drop-off Location (Ascending Order)')
plt.xlabel('Drop-off Location (index)')
plt.ylabel('Mean Distance')
plt.show()
```

**Histogram of rides by drop-off location**

First, check to whether the drop-off locations IDs are consecutively numbered. For instance, does it go 1, 2, 3, 4..., or are some numbers missing (e.g., 1, 3, 4...). If numbers aren't all consecutive, the histogram will look like some locations have very few or no rides when in reality there's no bar because there's no location. 

```python
# Check if all drop-off locations are consecutively numbered

unique_ids = sorted(df['DOLocationID'].unique())
print("First few IDs:", unique_ids[:10])
print("Last few IDs:", unique_ids[-10:])

# Check if IDs are consecutive
consecutive = all(unique_ids[i] + 1 == unique_ids[i + 1] for i in range(len(unique_ids) - 1))

if consecutive:
    print("✅ All drop-off location IDs are consecutive.")
else:
    missing = set(range(unique_ids[0], unique_ids[-1] + 1)) - set(unique_ids)
    print("⚠️ Some drop-off location IDs are missing:", sorted(missing))
```

This analysis shows that the drop-off location IDs are **not consecutively numbered**. Several IDs are missing from the sequence, which means that some potential drop-off zones do not appear in the dataset. This is normal in real-world data, where not all defined zones are necessarily used for trips.

To eliminate the spaces in the historgram that these missing numbers would create, sort the unique drop-off location values, then convert them to strings. This will make the histplot function display all bars directly next to each other. 

```python
#==> ENTER YOUR CODE HERE
# DOLocationID column is numeric, so sort in ascending order
sorted_ids = sorted(df['DOLocationID'].unique())

# Convert to string
sorted_ids_str = [str(i) for i in sorted_ids]

# Plot
plt.figure(figsize=(12, 3))
plt.plot(sorted_ids_str, np.ones_like(sorted_ids_str), '|', color='steelblue', markersize=10)
plt.title("Existing Drop-off Location IDs (No Gaps)")
plt.xlabel("Drop-off Location ID")
plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
# Plot a histogram of trip distances to check distribution and potential outliers

plt.figure(figsize=(8,4))
plt.hist(df['trip_distance'], bins=50, color='steelblue', edgecolor='black')
plt.title("Distribution of Trip Distances")
plt.xlabel("Trip Distance")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Create distance bins and calculate average fare per bin
bins = [0, 1, 2, 5, 10, 20, 50]
df['distance_bin'] = pd.cut(df['trip_distance'], bins=bins)

avg_fare_by_bin = df.groupby('distance_bin')['total_amount'].mean().reset_index()

# Plot
plt.figure(figsize=(8,4))
plt.bar(avg_fare_by_bin['distance_bin'].astype(str), avg_fare_by_bin['total_amount'], color='orange', edgecolor='black')
plt.title("Average Fare by Trip Distance Range")
plt.xlabel("Trip Distance Range (miles)")
plt.ylabel("Average Total Fare ($)")
plt.tight_layout()
plt.show()
```

- The bar chart shows a clear positive relationship between trip distance and total fare.
- As the distance increases, the average fare rises almost proportionally.
- This confirms that pricing generally scales with distance, as expected in a taxi fare system.
- However, the increase appears slightly non-linear for longer trips, which could suggest additional fees or time-based charges.
